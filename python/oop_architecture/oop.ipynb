ООП
├── Класс (class)
│   ├── атрибуты класса (species = "...")
│   ├── атрибуты экземпляра (self.name = ...)
│   └── методы
│       ├── обычный (def bark(self))
│       ├── @staticmethod — не знает ни о классе, ни об экземпляре
│       ├── @classmethod — знает о классе (cls), фабричный метод
│       └── @property — доступ как атрибут, но через метод
│           └── @x.setter — перехват присваивания + валидация
│
├── 4 кита
│   ├── Инкапсуляция
│   │   ├── __balance (name mangling → _Class__balance)
│   │   ├── _transactions (соглашение "не трогать")
│   │   └── @property скрывает реализацию
│   │
│   ├── Наследование
│   │   ├── class Cat(Animal) — одиночное
│   │   ├── super().__init__() — вызов родителя
│   │   └── isinstance / issubclass
│   │
│   ├── Полиморфизм
│   │   ├── утиная типизация — один интерфейс, разные объекты
│   │   ├── переопределение методов (speak в Cat)
│   │   └── перегрузка операторов (магические методы)
│   │
│   └── Абстракция
│       ├── ABC + @abstractmethod — нельзя создать напрямую
│       ├── @property @abstractmethod — абстрактное свойство
│       │   └── в подклассе ОБЯЗАН быть @property (diet)
│       └── raise NotImplementedError — "мягкая" абстракция (Shape)
│
├── Магические методы (dunder)
│   ├── __init__ — конструктор
│   ├── __str__ — для print()
│   ├── __repr__ — для отладки / repr()
│   ├── __add__, __sub__, __mul__, __rmul__ — арифметика
│   ├── __eq__, __lt__ + @total_ordering — сравнение
│   ├── __abs__, __neg__ — унарные операции
│   ├── __getitem__ — m[i][j]
│   ├── __len__ — len(obj)
│   ├── __iter__ — for x in obj
│   └── __call__ — obj() (используется в метаклассах)
│
├── Декораторы
│   ├── функциональные
│   │   ├── @timer — простой враппер
│   │   ├── @retry(n, delay) — декоратор с параметрами
│   │   │   └── functools.wraps — сохраняет __name__, __doc__
│   │   └── @cache(ttl)
│   │       ├── поддержка @cache и @cache(ttl=N)
│   │       │   └── callable(ttl) — различает два случая
│   │       ├── ключ кэша: (args, frozenset(kwargs.items()))
│   │       │   ├── args — tuple, хэшируемый
│   │       │   └── frozenset — неизменяемое множество, хэшируемый
│   │       ├── значение: (result, timestamp)
│   │       ├── nonlocal — доступ к счётчикам из внешней области
│   │       └── .cache_info() / .cache_clear() — атрибуты функции
│   └── классовые
│       ├── @staticmethod
│       ├── @classmethod
│       └── @property / @x.setter
│
├── Метаклассы
│   ├── type — метакласс по умолчанию
│   │   └── type(name, bases, namespace) — создание класса вручную
│   ├── SingletonMeta(type)
│   │   └── __call__ — перехват создания экземпляра
│   ├── PluginMeta(ABCMeta)
│   │   └── __new__ — перехват создания класса, авторегистрация
│   └── ModelMeta(type) [ORM]
│       └── __new__ — сбор Field-дескрипторов в _fields
│
└── Типы данных и хэшируемость
    ├── хэшируемые (могут быть ключом dict / элементом set)
    │   ├── int, float, str, bool
    │   ├── tuple — если все элементы хэшируемые
    │   └── frozenset
    └── не хэшируемые (изменяемые)
        ├── list → tuple для хэширования
        ├── dict → frozenset(items()) для хэширования
        └── set → frozenset
