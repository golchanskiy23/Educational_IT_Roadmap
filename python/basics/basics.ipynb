```
Python Basics
│
├── Переменные и типы
│   ├── динамическая типизация — тип определяется значением, не объявлением
│   │   └── x = 42 → x = 'hello' → x = [1,2,3] — всё валидно
│   ├── множественное присваивание
│   │   ├── a, b = 3, 4
│   │   ├── a, b = b, a — обмен без temp
│   │   └── x = y = z = 0
│   └── константы — соглашение (UPPER_CASE), не ограничение
│
├── Типы данных
│   ├── Числа
│   │   ├── int — целое
│   │   ├── float — дробное
│   │   └── bool — подкласс int (True=1, False=0)
│   │       ├── bool(0) → False
│   │       ├── bool('') → False (пустая строка)
│   │       └── bool(-5) → True (любое ненулевое)
│   │
│   ├── str — неизменяемая последовательность символов
│   │   ├── f-строки: f'{val:<{width}.2f}'
│   │   ├── split() → list[str]
│   │   ├── join(iterable) → str
│   │   └── ISO-дата сортируется лексикографически
│   │
│   ├── list — изменяемая последовательность
│   │   ├── индексы: lst[0], lst[-1]
│   │   ├── срезы: lst[1:3], lst[::-1]
│   │   ├── методы: append, insert, remove, pop, sort
│   │   ├── list comprehension: [f(x) for x in xs if cond]
│   │   └── ЛОВУШКА: [[0]*m]*n — все строки один объект
│   │       └── fix: [[0]*m for _ in range(n)]
│   │
│   ├── tuple — неизменяемая последовательность
│   │   ├── single = (42,) — нужна запятая!
│   │   ├── распаковка: x, y, z = (1, 2, 3)
│   │   ├── *rest: first, *rest = (1,2,3,4)
│   │   ├── хэшируемый → может быть ключом dict
│   │   └── namedtuple — кортеж с именами полей
│   │
│   ├── dict — хэш-таблица ключ→значение
│   │   ├── d.get(key, default) — безопасное чтение
│   │   ├── d.items() / .keys() / .values()
│   │   ├── del d[key] / d.pop(key)
│   │   ├── d1 | d2 — слияние (Python 3.9+)
│   │   ├── dict comprehension: {x: x**2 for x in range(5)}
│   │   └── применение: group_by, count, ops-таблица
│   │
│   └── set — неупорядоченное множество уникальных элементов
│       ├── empty_set = set() — НЕ {} (это dict!)
│       ├── add, remove, discard
│       ├── | & - ^ — объединение, пересечение, разность
│       ├── быстрая проверка: x in s — O(1) vs O(n) у list
│       └── list(set(lst)) — убрать дубли
│
├── Копирование (критически важно для изменяемых типов)
│   ├── = — присваивание ссылки (a is b → True)
│   ├── shallow copy — новый контейнер, те же вложенные объекты
│   │   ├── lst.copy() / list(lst) / lst[:]
│   │   └── ЛОВУШКА: вложенные списки всё ещё общие
│   └── deep copy — рекурсивное копирование всего
│       └── copy.deepcopy(obj)
│           └── используется в adjacent_matrix для независимой копии
│
├── Управление потоком
│   ├── if / elif / else
│   ├── тернарный оператор: x if cond else y
│   ├── and / or / not
│   ├── while + break / continue
│   └── for
│       ├── for x in iterable
│       ├── for i in range(start, stop, step)
│       ├── for i, x in enumerate(lst)
│       └── for k, v in d.items()
│
├── Матрицы
│   ├── список списков: matrix[i][j]
│   ├── матрица смежности: edges → matrix[i][j] = matrix[j][i] = 1
│   ├── степень вершины: sum(matrix[i])
│   └── numpy: A @ B (умножение), A.T (транспонирование)
│
├── Файловая система
│   ├── os модуль
│   │   ├── os.getcwd() — текущая директория
│   │   ├── os.listdir(path) — содержимое (первый уровень)
│   │   ├── os.path.join(dir, file) — кроссплатформенный путь
│   │   ├── os.path.isfile / isdir / exists
│   │   └── os.path.getsize(path) — размер в байтах
│   ├── pathlib.Path — ООП-обёртка
│   │   ├── p.parent / .name / .stem / .suffix
│   │   ├── p.iterdir() — содержимое директории
│   │   └── p.glob('*.py') — поиск по паттерну
│   └── os.environ.get('HOME') — переменные окружения
│
└── Паттерны из кода
    ├── group_by через dict: cat → [names]
    ├── count через dict.get(key, 0) + 1
    ├── max(dict, key=lambda k: dict[k]) — ключ с макс. значением
    ├── sorted([...], key=lambda x: x['field']) — сортировка по полю
    └── assert — проверка инварианта (оригинал не тронут)
```
