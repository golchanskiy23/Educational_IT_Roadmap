```
Python Advanced
│
├── Генераторы и итераторы
│   ├── Итерируемый (Iterable)
│   │   └── имеет __iter__ → возвращает итератор
│   │       └── list, str, dict, set, tuple
│   │
│   ├── Итератор (Iterator)
│   │   ├── имеет __iter__ + __next__
│   │   ├── хранит состояние (текущая позиция)
│   │   └── class Countdown — ручная реализация
│   │
│   ├── Генератор — ленивый итератор
│   │   ├── функция с yield — приостанавливается, отдаёт значение
│   │   ├── генераторное выражение: (x**2 for x in range(n))
│   │   │   └── vs list comprehension [x**2 ...] — всё в памяти сразу
│   │   ├── yield from — перепоручает перебор вложенному итератору
│   │   │   └── используется в рекурсивном all_files()
│   │   └── itertools.islice — срез ленивой последовательности
│   │
│   └── Рекурсия vs Генератор
│       ├── рекурсия: собирает весь стек вызовов → результат
│       └── генератор: один элемент → отдаёт → забывает → следующий
│
├── Исключения
│   ├── Иерархия
│   │   └── BaseException
│   │       ├── SystemExit, KeyboardInterrupt (ctrl+C)
│   │       └── Exception
│   │           ├── ValueError
│   │           │   └── CalcError (пользовательское)
│   │           ├── ZeroDivisionError
│   │           ├── TypeError
│   │           ├── FileNotFoundError
│   │           └── пользовательские: наследуем от Exception
│   │               ├── StackError
│   │               │   ├── StackEmptyError
│   │               │   └── StackFullError (доп. атрибут limit)
│   │               └── InsufficientFundsError (balance, amount)
│   │
│   ├── try / except / else / finally
│   │   ├── except — перехват конкретного типа
│   │   ├── else — выполняется если исключений НЕ было
│   │   └── finally — выполняется ВСЕГДА (даже при return)
│   │
│   └── raise / reraise / raise from
│       ├── raise MyError(...) — поднять своё
│       ├── raise — без аргументов, пробросить то же дальше
│       └── raise NewError() from original — цепочка причин
│
├── Модули и пакеты
│   ├── модуль — любой .py файл
│   ├── пакет — папка с __init__.py
│   │   └── mypackage/geometry.py → from mypackage import geometry
│   ├── __pycache__ — байт-код (.pyc), создаётся автоматически
│   ├── pip — установка из PyPI
│   └── venv — изолированное окружение
│       ├── python -m venv .venv
│       ├── PATH подменяется при активации
│       └── site-packages изолированы между проектами
│
├── Структуры данных (Stack)
│   ├── LIFO — последним вошёл, первым вышел
│   ├── магические методы протокола
│   │   ├── __len__, __bool__, __str__, __repr__
│   │   ├── __iter__, __contains__, __eq__
│   │   └── push / pop / top — основные операции
│   └── применение: ОПН (обратная польская нотация)
│       └── evaluate("2 3 + 1 5 * 2 / -")
│           └── использует ops dict: str → operator.func
│
├── Типы данных и хэшируемость
│   ├── хэшируемые → могут быть ключом dict / элементом set
│   │   ├── int, float, str, bool, None
│   │   ├── tuple (если все элементы хэшируемые)
│   │   └── frozenset
│   └── не хэшируемые (изменяемые)
│       ├── list → tuple(...)
│       ├── dict → frozenset(d.items())
│       └── set → frozenset(...)
│
└── Паттерны из кода
    ├── operator модуль — функции вместо операторов
    │   └── ops = {"+": operator.add, ...} → ops["+"](a, b)
    ├── pathlib.Path — ООП-обёртка над os.path
    │   ├── path.iterdir() — содержимое директории
    │   ├── entry.is_file() / is_dir()
    │   └── entry.stat().st_size — размер файла
    └── defaultdict — dict с дефолтным значением для новых ключей
```
