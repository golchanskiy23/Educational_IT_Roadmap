Функции Python
│
├── Основы функций
│   ├── аргументы
│   │   ├── позиционные (a, b)
│   │   ├── именованные с дефолтом (greeting='Hello')
│   │   │   └── ЛОВУШКА: изменяемый дефолт (lst=[])
│   │   │       └── создаётся ОДИН РАЗ при определении функции
│   │   │           └── fix: def f(lst=None): if lst is None: lst = []
│   │   ├── *args — кортеж позиционных
│   │   └── **kwargs — словарь именованных
│   │
│   ├── области видимости (LEGB)
│   │   ├── Local — внутри функции
│   │   ├── Enclosing — внешняя функция (замыкание)
│   │   ├── Global — модуль
│   │   └── Built-in — встроенные (len, print...)
│   │       ├── global counter — изменить глобальную
│   │       └── nonlocal count — изменить из enclosing
│   │
│   └── передача аргументов
│       ├── int, str, float — передаются по значению (копия)
│       └── list, dict — передаются по ссылке (мутируют оригинал)
│
├── Функции высшего порядка
│   ├── принимают функцию как аргумент
│   │   ├── map(func, iterable) → ленивый итератор
│   │   ├── filter(func, iterable) → ленивый итератор
│   │   └── sorted(iterable, key=func)
│   ├── возвращают функцию
│   │   └── make_multiplier(a) → multiply(b)
│   └── functools.reduce(func, iterable)
│       └── ((((1+2)+3)+4)+5) — свёртка
│
├── Lambda
│   ├── анонимная функция: lambda x: x**2
│   ├── используется как key= в sorted/max/min
│   └── classify = lambda avg: 'отлично' if avg >= 85 else ...
│       └── тернарный оператор цепочкой
│
├── Замыкания (Closure)
│   ├── внутренняя функция захватывает переменную из enclosing
│   ├── make_counter() → increment() — счётчик через nonlocal
│   ├── make_multiplier(a) → multiply(b) — фабрика функций
│   └── make_level_filter(level) → lambda entry: entry['level'] == level
│       └── применение: filter(make_level_filter('ERROR'), entries)
│
├── Работа с файлами
│   ├── open(path, mode, encoding)
│   │   ├── 'r' — чтение, 'w' — запись, 'a' — дополнение
│   │   └── with open(...) as f — автозакрытие (контекстный менеджер)
│   ├── f.read() — всё содержимое → str
│   ├── f.write(str) — запись строки
│   └── os.path.join / pathlib.Path — кроссплатформенные пути
│       └── __file__ — путь к текущему модулю
│
├── Строки и форматирование
│   ├── f-строки: f'{value:<{width}}' — выравнивание по ширине
│   │   ├── < — по левому краю
│   │   ├── > — по правому краю
│   │   └── .2f — два знака после запятой
│   ├── str.split() — разбить по пробелам
│   ├── ' '.join(iterable) — собрать строку
│   └── str.strip() — убрать пробелы/переносы
│
├── datetime
│   ├── datetime.strptime(str, fmt) — строка → объект
│   ├── datetime - datetime → timedelta
│   └── timedelta.total_seconds() → float
│
├── Типы данных в контексте функций
│   ├── dict
│   │   ├── d.get(key, default) — безопасное чтение
│   │   ├── d.items() → пары (key, value)
│   │   └── используется как: ops = {"+": operator.add}
│   ├── list
│   │   ├── list comprehension: [f(x) for x in xs if cond]
│   │   └── list(map(...)) / list(filter(...)) — материализация
│   ├── tuple
│   │   └── строки таблицы: ('Алиса', 'Инженер', 95000)
│   └── str
│       └── ISO-формат даты сортируется лексикографически
│           └── min/max по строкам "YYYY-MM-DD HH:MM:SS" корректен
│
└── Паттерны из кода
    ├── format_cell — isinstance(value, int) → разное выравнивание
    ├── zip(headers, col_widths) — параллельный обход двух списков
    ├── enumerate(iterable, start=1) — индекс + значение
    ├── operator модуль — ops[token](a, b) вместо if/elif цепочки
    └── level_counts.get(level, 0) + 1 — счётчик без KeyError
